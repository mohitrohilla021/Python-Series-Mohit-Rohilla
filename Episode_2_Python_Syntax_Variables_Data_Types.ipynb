## 3. Python Data Types
Python has built-in data types like:
- `int`
- `float`
- `str`
- `bool`
- `list`
- `tuple`
- `dict`
- `set`

```python
# Datatype examples
x_int = 10       # int
y_float = 3.14   # float
name = "Python"  # string
is_active = True # boolean
```

```python
# Checking type
print(type(x_int), type(y_float), type(name), type(is_active))
```

### Collections

```python
lst = [1,2,3]
tup = (1,2,3)
dct = {"name":"Python","year":1991}
st = {1,2,3}
print(lst, tup, dct, st)
```
