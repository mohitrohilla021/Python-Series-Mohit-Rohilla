## 1. What is a String?
A string is a sequence of characters enclosed in single or double quotes.

```python
name = "Mohit Rohilla"
company = "ABC"
name, company
```

## 2. Creating Strings

```python
a = "Hello"
b = "Python is amazing"
c = """This is a 
multi-line string."""
a, b, c
```

## 3. Accessing Characters (Indexing)

```python
word = "Python"
print(word[0])
print(word[3])
print(word[-1])
```

## 4. Slicing Strings

```python
text = "Programming"
print(text[0:6])
print(text[3:])
print(text[:5])
print(text[-5:])
```

## 5. Length of a String

```python
msg = "Hello World"
len(msg)
```

## 6. String Methods

### Changing Case

```python
s = "python programming"
s.upper(), s.lower(), s.title(), s.capitalize()
```

### Removing Spaces

```python
data = "   hello python   "
data.strip(), data.lstrip(), data.rstrip()
```

### Replace Text

```python
msg = "I love Java"
msg.replace("Java", "Python")
```

### Count Occurrence

```python
text = "banana"
text.count("a")
```

### Find Substrings

```python
line = "Python is awesome"
line.find("is"), line.find("java")
```

### Split & Join

```python
sentence = "python is super powerful"
words = sentence.split(" ")
joined = "-".join(words)
words, joined
```

## 7. String Operations

### Concatenation

```python
first = "Data"
second = "Science"
first + " " + second
```

### Repetition

```python
"Hi! " * 3
```

## 8. Escape Characters

```python
print("He said \"Python is easy!\"")
print("Line1\nLine2")
print("Tab\tSpace")
```

## 9. f-Strings (Modern Python Formatting)

```python
name = "Mohit"
age = 25
f"My name is {name} and I am {age} years old."
```

## 10. Real-Life Example: Cleaning Email Input

```python
email = " Rohilla@gmail.com  "
clean_email = email.strip().lower()
clean_email
```
