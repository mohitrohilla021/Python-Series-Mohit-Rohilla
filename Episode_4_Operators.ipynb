## 1Ô∏è‚É£ Arithmetic Operators
Used for mathematical calculations.

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

## 2Ô∏è‚É£ Assignment Operators
Used to assign and update values.

```python
x = 10
print("Initial:", x)

x += 5
print("After += :", x)

x -= 3
print("After -= :", x)

x *= 2
print("After *= :", x)

x /= 4
print("After /= :", x)
```

## 3Ô∏è‚É£ Comparison Operators
Used to compare two values. Result is always True or False.

```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

## 4Ô∏è‚É£ Logical Operators
Used to combine conditional statements.

```python
x = 10
y = 20

print(x > 5 and y > 15)
print(x > 15 or y > 15)
print(not(x > 5))
```

## 5Ô∏è‚É£ Identity Operators
Used to compare memory locations.

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)
print(a is c)
print(a is not c)
```

## 6Ô∏è‚É£ Membership Operators
Used to test if a value exists in a sequence.

```python
fruits = ["apple", "banana", "mango"]

print("apple" in fruits)
print("grape" not in fruits)
```

## 7Ô∏è‚É£ Bitwise Operators
Operate on binary numbers (advanced but important).

```python
a = 5
b = 3

print("AND:", a & b)
print("OR:", a | b)
print("XOR:", a ^ b)
print("Left Shift:", a << 1)
print("Right Shift:", a >> 1)
```

## 8Ô∏è‚É£ Operator Precedence
Python follows a specific order while evaluating expressions.

```python
print(10 + 5 * 2)
print((10 + 5) * 2)
```

## üî• Real-Life Example ‚Äî Login Validation

```python
username = "admin"
password = "1234"

input_user = "admin"
input_pass = "12345"

print(input_user == username and input_pass == password)
```

## üî• Real-Life Example ‚Äî Even or Odd Checker

```python
num = 15

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```
