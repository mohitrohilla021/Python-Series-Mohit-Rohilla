# Video 5 ‚Äî Conditional Statements in Python (if / elif / else)
---
Conditional statements help Python make decisions just like humans do.

## 1Ô∏è‚É£ What are Conditional Statements?
They allow your program to execute code based on conditions.

Python uses:
- if
- elif
- else

## 2Ô∏è‚É£ Basic `if` Statement

```python
age = 20

if age >= 18:
    print("You are eligible to vote")
```

## 3Ô∏è‚É£ `if-else` Statement

```python
age = 16

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

## 4Ô∏è‚É£ `if-elif-else` Statement

```python
marks = 30

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

## 5Ô∏è‚É£ Multiple Conditions using Logical Operators

```python
age = 25
has_id = True

if age >= 18 and has_id:
    print("Entry allowed")
else:
    print("Entry denied")
```

## 6Ô∏è‚É£ Nested `if` Statements

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Invalid user")
```

## 7Ô∏è‚É£ Real-Life Example ‚Äî Traffic Signal

```python
signal = "red"

if signal == "red":
    print("Stop")
elif signal == "yellow":
    print("Get Ready")
elif signal == "green":
    print("Go")
else:
    print("Invalid signal")
```

## 8Ô∏è‚É£ Real-Life Example ‚Äî ATM Withdrawal

```python
balance = 5000
withdraw = 3000

if withdraw <= balance:
    balance -= withdraw
    print("Withdraw successful")
    print("Remaining balance:", balance)
else:
    print("Insufficient balance")
```

## 9Ô∏è‚É£ Ternary Conditional (One-line if-else)

```python
age = 20
result = "Adult" if age >= 18 else "Minor"
print(result)
```
